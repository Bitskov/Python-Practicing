```python
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 1000);
```

```python
plt.plot(x, np.sin(x));
plt.title("A Sine Curve");
plt.xlabel("x");
plt.ylabel("y=sin(x)");
```

```python
plt.plot(x, np.sin(x), '-r', label='sin(x)');
plt.plot(x, np.cos(x), ':b', label='cos(x)');
plt.axis('equal');
plt.legend();
```
