```python
import numpy as np
%matplotlib inline
import matplotlib as mtl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x));
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '-');
plt.plot(x, np.cos(x), '--');
```

```python
plt.figure();
plt.subplot(2, 1, 1);
plt.plot(x, np.sin(x));
plt.subplot(2, 1, 2);
plt.plot(x, np.cos(x));
```

```python
fig, ax = plt.subplots(2);
ax[0].plot(x, np.sin(x));
ax[1].plot(x, np.cos(x));
```

```python
plt.style.use('seaborn-whitegrid');
ax = plt.axes();
ax.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x - 0), color='blue');
plt.plot(x, np.sin(x - 1), color='g'); # Color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75'); # From white to black colour (1 -> 0)
plt.plot(x, np.sin(x - 3), color='#FFDD44');
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)); # RGB tuple (0 -> 1)
plt.plot(x, np.sin(x - 5), color='chartreuse'); # Colors from HTML
```

```python
plt.plot(x, x + 0, linestyle='solid'); 
plt.plot(x, x + 1, linestyle='dashed');
plt.plot(x, x + 2, linestyle='dashdot');
plt.plot(x, x + 3, linestyle='dotted');
```

```python
plt.plot(x, x + 0, '-'); # Analog of upper linestyles
plt.plot(x, x + 1, '--');
plt.plot(x, x + 2, '-.');
plt.plot(x, x + 3, ':');
```

```python
plt.plot(x, x + 0, '-g'); # Colors can be added to linestyle
plt.plot(x, x + 1, '--c');
plt.plot(x, x + 2, '-.k');
plt.plot(x, x + 3, ':r');
```
